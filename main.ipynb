```python
import pandas as pd
import os
from sqlalchemy import create_engine
import logging
import time

logging.basicConfig(
    filename="logs/ingestion_db.log", 
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s", 
    filemode="a"  
)
engine = create_engine('sqlite:///inventory.db')
def ingest_db(df,table_name, engine):
    df.to_sql(table_name, con=engine, if_exists= 'replace', index=False)
def load_raw_data():
    start = time.time()
    for file in os.listdir('data'):
        if file.endswith('.csv'):
            file_path = 'data/' + file
            table_name = file[:-4]

            try:
                if file == 'sales.csv':
                    # Special handling for big file
                    first = True
                    for chunk in pd.read_csv(file_path, chunksize=5000):
                        chunk.to_sql(table_name, con=engine, if_exists='replace' if first else 'append', index=False)
                        first = False
                        print(f'Inserted {len(chunk)} rows into {table_name}')
                    logging.info(f'✅ Ingested {file} in chunks')
                else:
                    df = pd.read_csv(file_path)
                    df.to_sql(table_name, con=engine, if_exists='replace', index=False)
                    logging.info(f'✅ Ingested {file}')
            except Exception as e:
                logging.error(f'❌ Failed to ingest {file}: {e}')
    end = time.time()
    total_time = (end - start)/60
    logging.info('--------------Ingestion Complete------------')
    logging.info(f'\nTotal Time Taken: {total_time} minutes')

if __name__ == '__main__':
    load_raw_data()
```
